```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def step(board):
    new_board = np.zeros(board.shape)
    rows = board.shape[0]
    cols = board.shape[1]
    for row in range(rows):
        for col in range(cols):
            neighbors = [
                (row - 1, col - 1),
                (row - 1, col),
                (row - 1, col + 1),
                
                (row, col - 1),
                (row, col + 1),
                
                (row + 1, col - 1),
                (row + 1, col),
                (row + 1, col + 1),
            ]
            num_alive = 0
            for neighbor in neighbors:
                n_row, n_col = neighbor
                if 0 <= n_row < rows and 0 <= n_col < cols and board[n_row][n_col]:
                    num_alive += 1
            if board[row][col]:
                if num_alive >= 2:
                    new_board[row][col] = True
            else:
                if num_alive >= 3:
                    new_board[row][col] = True
    return new_board

def plot_board(board):
    plt.imshow(board, cmap='hot')
    plt.show()
```

```python
size = 100
board = np.zeros((size, size))
board[50, 50] = True
board[51, 50] = True
board[52, 50] = True

time_steps = 10
while True:
    plot_board(board)
    board = step(board)
```
